```python
# Dependencies
import numpy as np
import pandas as pd
import os
```

```python
enron = pd.read_csv(os.path.join('enron_salary_data.csv'))
enron_df = pd.DataFrame(enron)
enron_df.head()
```

```python
enron_df = enron_df[enron_df.name != 'TOTAL']
```

```python
enron_df.name.count()
```

```python
drop_columns = ['name', 'email_address']
enron_df.drop(drop_columns, inplace=True, axis=1)
```

```python
enron_df.head()
```

```python
enron_df.info()
```

```python
#Keeps only columns that have ~70% of the fields as non-null
slim_enron_df = enron_df.dropna(thresh=14)
slim_enron_df.info()
```

```python
slim_enron_df.to_csv('enron_test_data.csv', index=False)
```
