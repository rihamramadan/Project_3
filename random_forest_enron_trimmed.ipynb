```python
from sklearn import tree
import pandas as pd
import os
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(os.path.join("enron_test_data.csv"))
enron_df = pd.DataFrame(df)
enron_df.head()
```

```python
for column in df.columns:
    df[column] = df[column].apply(lambda x: np.NaN if x == 'NaN' else x)
```

```python
enron_df.fillna(0,inplace=True)
enron_df.head()
```

```python
target = enron_df["poi"]
target_names = ["False", "True"]
```

```python
data = enron_df.drop("poi", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(f"Decision Tree Classifier Score: {clf.score(X_test, y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
print(f"Random Forest Tree Classifier Score: {rf.score(X_test, y_test)}")
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
