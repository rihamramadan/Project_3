```python
# Dependencies
import numpy as np
import pandas as pd
import os
```

```python
enron = pd.read_csv(os.path.join('enron_salary_data.csv'))
enron_df = pd.DataFrame(enron)
enron_df.head()
```

```python
enron_numeric_df = enron_df
enron_numeric_df.dtypes
```

```python
for column in enron_numeric_df.columns:
    enron_numeric_df[column] = enron_numeric_df[column].apply(lambda x: np.NaN if x == 'NaN' else x)

enron_numeric_df.info()
```

```python
enron_df.head()
```

```python
enron_numeric_df.fillna(0,inplace=True)
enron_numeric_df.info()
```

```python
enron_numeric_df.to_csv('enron_salary_numeric.csv', index=False)
```
