```python
# Drop Total Column
enron_df = enron_df[enron_df.name != 'TOTAL']

enron_df = enron_df[enron_df.name != 'THE TRAVEL AGENCY IN THE PARK']
```

```python
# Drop non numeric columns
drop_columns = ['name', 'email_address']
enron_df.drop(drop_columns, inplace=True, axis=1)
enron_df.head()
```

```python
target = enron_df["poi"]
target_names = ["False", "True"]
```

```python
data = enron_df.drop("poi", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
print(f"Decision Tree Classifier Score: {clf.score(X_test, y_test)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
print(f"Random Forest Tree Classifier Score: {rf.score(X_test, y_test)}")
```
